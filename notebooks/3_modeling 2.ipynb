Creating a threshold where we are only including restaurants where review_count is greater than 100. 

```python
final_data = final_data[final_data['review_count'] >= 100]
final_data
```

Creating a threshold where we are only including users where review count is greater than 50. 

```python
# Group the 'final_data' DataFrame by 'user_id' and count the number of ratings/stars given by each user.
user_ratings_count = final_data.groupby('user_id').count()['stars']

# Create a boolean Series 'avid_reviewers', which indicates whether each user has rated more than 200 times or not.
high_reviewers = user_ratings_count > 50

# Filter the 'user_ratings_count' Series to get only the index (user_id) of users who have rated more than 200 times.
restaurant_reviewers = user_ratings_count[high_reviewers].index
```

To ensure the users with more than 50 reviews are included in the dataframe, we check using isin().

```python
final_data_ratings = final_data[final_data['user_id'].isin(restaurant_reviewers)]
final_data_ratings.shape
```

### Collaborative-Filtering Recommendation System

```python
final_data_ratings_pivot = final_data_ratings.pivot_table(index='restaurant_name', columns='user_id', values='stars')
```

```python
final_data_ratings_pivot.shape
```

```python
# Displaying the first few rows to get an initial glimpse of the data
final_data_ratings_pivot.head()
```

```python
final_data_ratings_pivot.fillna(0, inplace=True)
final_data_ratings_pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_scores = cosine_similarity(final_data_ratings_pivot)
similarity_scores.shape
```

```python
def recommend(restaurant_name):
    # Find the index of the input restaurant name in the pivot table
    index = np.where(final_data_ratings_pivot.index == restaurant_name)[0][0]

    # Retrieve the similarity scores of the input restaurant with other restaurants,
    # sort them in descending order, and select the top 4 similar items
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:5]

    # Initialize an empty list to store recommended restaurant names
    data = []

    # Iterate through each similar item
    for i in similar_items:
        # Fetch the relevant restaurant name from the 'business_data' dataset
        similar_restaurant_name = final_data_ratings_pivot.index[i[0]]

        # Append the restaurant name to the 'data' list
        data.append(similar_restaurant_name)

    # Return the 'data' list containing names of the recommended restaurants
    return data
```

```python
# Filter data for the city of Vancouver
vancouver_data = final_data[final_data['city'] == 'Vancouver']

# Sort the filtered data by the 'review_count' column in descending order
vancouver_data_sorted = vancouver_data.sort_values('review_count', ascending=False)

# Drop duplicates of 'restaurant_name' to get unique restaurants
vancouver_data_unique = vancouver_data_sorted.drop_duplicates(subset='restaurant_name')

# Display the top 5 restaurants with the highest review_count
top_10_restaurants = vancouver_data_unique.head(10)
top_10_restaurants[['restaurant_name', 'address', 'city', 'state', 'restaurant_rating', 'review_count', 'categories']]
```
